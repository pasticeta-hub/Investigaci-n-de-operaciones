# Metodo implicito de Richardson

En esta tarea vamos a resolver de forma numerica la ecuacion del calor 1D en el intervalo 0 < x < 1 con condiciones de frontera tipo Dirichlet:

- en los extremos de la barra la temperatura vale cero todo el tiempo
- al inicio del tiempo la temperatura tiene forma de seno

La ecuacion continua que queremos aproximar es:

- ecuacion del calor:  u_t = u_xx
- condicion inicial:   u(x,0) = sin(pi * x)
- fronteras:           u(0,t) = 0,  u(1,t) = 0

Para resolver esto de forma numerica hacemos una malla:

- paso en x: h
- paso en tiempo: k
- nodos interiores en x y nodos en t

El metodo implicito (tipo Richardson) lo que hace es que, para ir de un tiempo n al tiempo n+1, tenemos que resolver un sistema lineal.

Si se escribe la ecuacion en diferencias hacia adelante en tiempo e implicita en espacio, queda algo de este estilo (solo como idea):

( u_i^{n+1} - u_i^n ) / k  â‰ˆ  ( u_{i+1}^{n+1} - 2 u_i^{n+1} + u_{i-1}^{n+1} ) / h^2

Reacomodando se puede dejar todo lo de tiempo n+1 del lado izquierdo y u_i^n del lado derecho. Eso se resume en

A * u^{n+1} = u^n

donde A es una matriz tridiagonal (3 en la diagonal y -1 en las diagonales vecinas) cuando tomamos r = k / h^2 = 1.

En resumen: en cada paso de tiempo resolvemos el sistema A * u_nuevo = u_viejo y asi vamos viendo como se enfria la barra.


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm  # colormap

# condicion inicial
def f(x):
    return np.sin(np.pi * x)
```

```python
# queremos una matriz 9x9 con 3 en la diagonal y -1 en las diagonales vecinas

n = 9

# creeamos una matriz de ceros y luego la vamos llenando
A = sp.zeros(n, n)

for i in range(n):
    A[i, i] = 3           # diagonal principal
    if i > 0:
        A[i, i-1] = -1    # debajo de la diagonal
    if i < n - 1:
        A[i, i+1] = -1    # arriba de la diagonal

#ya que lit nuestra matriz como en escalerita repitiendose siempre podemos usareste ciclo para que se llene solita
A
```

```python
# nodos interiores donde usamos la condicion inicial: 0.1, 0.2, ..., 0.9
xs = np.linspace(0.1, 0.9, 9)

lista_b = []
for x in xs:
    lista_b.append(f(x))

# ahora pasamoos la lista a vector columna de sympy
b = sp.Matrix(lista_b)

b
```

     aplicamos el esquema j veces y nos regresa solo el vector final
    A -> matriz del sistema
    b -> vector inicial
    j -> numero de pasos en tiempo

```python
def richardson_1(A, b, j):
    b_1 = b.copy()
    for i in range(j + 1):
        sol = A.LUsolve(b_1)
        b_1 = sol

    return b_1
richardson_1(A, b, 3)
```

    aqui queremos guardar la evolucion completa.
    cada fila del resultado es el vector de temperaturas en un tiempo distinto.

```python
def richardson(A, b, j):

    S = []          # aqui voy guardando las soluciones
    b_1 = b.copy()  # copiamos para no modificar b original

    for i in range(j + 1):
        # guardamos el vector actual como fila (lo paso a numpy)
        S.append(np.array(b_1, dtype=float).reshape(-1))
        # resolvemos el sistema para el siguiente tiempo
        sol = A.LUsolve(b_1)
        b_1 = sol

    # regresa un arreglo numpy donde S[0] es t=0, S[1] es el siguiente tiempo, etc.
    return np.array(S)
```

```python
j = 20   # numero de pasos en el tiempo

U = richardson(A, b, j)
U.shape   # deberia ser (j+1, 9)
```

```python
# solucion exacta para este problema:
# u(x,t) = exp(-pi^2 t) * sin(pi x)

def u_exacto(x, t):
    return np.exp(-(np.pi**2) * t) * np.sin(np.pi * x)

# escogemos un t0 dentro del intervalo [0, 0.6]
t0 = 0.3

# paso de tiempo k (el profe usa 0.6 como tiempo final)
k = 0.6 / j

# indice aproximado que corresponde a t0
indice_t0 = int(t0 / k)

# nodos interiores donde tenemos la solucion numerica
x_internos = np.linspace(0.1, 0.9, 9)

u_num = U[indice_t0, :]             # solucion numerica en t0
u_ex = u_exacto(x_internos, t0)     # solucion exacta en t0

plt.figure()
plt.plot(x_internos, u_num, 'o-', label='numerica richardson')
plt.plot(x_internos, u_ex, 's--', label='exacta')
plt.xlabel('x')
plt.ylabel('u(x, t0)')
plt.title(f'comparacion en t = {t0:.2f}')
plt.legend()
plt.grid(True)
plt.show()
```

en esta grafica fijamos un tiempo t0 = 0.3.

- con circulitos se ve la solucion numerica que sale del metodo de richardson.
- con cuadritos se ve la formula exacta u(x,t) = exp(-pi^2 t) * sin(pi x).

la idea es comparar las dos curvas en los puntos interiores de la barra.  
se nota que casi se empalman, lo cual indica que con esta malla el metodo
esta aproximando bastante bien la solucion exacta.


**ejercicio 5. Â¿de verdad necesitamos el sistema de 9 x 9?**

en este problema la condicion inicial es

u(x,0) = sin(pi x)

y esta funcion es un modo propio tipico de la ecuacion del calor con
condiciones de dirichlet en 0 y 1. por teoria se sabe que la solucion exacta
tiene la forma

u(x,t) = exp(-pi^2 t) * sin(pi x)

o sea, la parte que depende de x no cambia: siempre es el mismo seno,
lo unico que se mueve es el factor exp(-pi^2 t) que va bajando.

por eso, en teoria podriamos trabajar solo con una ecuacion para ese coeficiente
y no montar todo el sistema 9 x 9. sin embargo, aqui si armamos la matriz A y
el sistema completo porque la idea del ejercicio es practicar el metodo
implicito, el armado de la matriz tridiagonal y ver como se ve la solucion en
la malla.


```python
# aqui el profe toma x desde 0 hasta 1 con 9 nodos,
# pero la solucion numerica esta en 0.1, 0.2, ..., 0.9
# esto es justo lo que comentamos en el ejercicio 6

x = np.linspace(0, 1, 9)        # nodos para graficar
y = np.linspace(0, 0.6, j + 1)  # tiempos
X, Y = np.meshgrid(x, y)

plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, U, levels=25, cmap=cm.viridis)
plt.colorbar(contour)
plt.title('solucion numerica de un problema de dirichlet')
plt.xlabel('x')
plt.ylabel('t')
plt.axis('equal')
plt.show()
```

en este mapa de colores:

- el eje horizontal es la posicion x en la barra,
- el eje vertical es el tiempo t,
- cada color representa un rango de valores de u(x,t).

al inicio (t cerca de 0) se ve una zona con valores mas altos en el centro,
por la forma de seno. conforme pasa el tiempo los colores intensos se van
apagando hasta que todo se acerca a cero. eso significa que el calor se va
perdiendo porque en los extremos de la barra mantenemos la temperatura en cero.


```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, U, cmap=cm.viridis, edgecolor='k')
ax.set_title('solucion de la ecuacion del calor con dirichlet')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('u(x,t)')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

esta grafica muestra la misma informacion pero en 3d:

- en el plano de abajo (x, t) se ve la posicion y el tiempo,
- la altura de la superficie es el valor de u(x,t).

al principio la superficie tiene una "loma" por la forma de seno,
y luego esa loma se va aplastando hasta casi pegarse al plano u = 0.
asi se ve visualmente como el calor se va perdiendo con el tiempo.


**ejercicio 6. ligero detalle con la precision de las graficas**

hay un detalle importante:

- la solucion numerica se calculo en los nodos interiores 0.1, 0.2, ..., 0.9
  (no usamos 0 ni 1 al resolver el sistema).
- pero al graficar usamos `x = np.linspace(0, 1, 9)`, que incluye los extremos 0 y 1.

eso quiere decir que estamos dibujando los valores numericos como si
pertenecieran a otra malla. en otras palabras, los puntos se mueven un poquito
en el eje x. como matplotlib interpola suave, el error visual casi no se nota,
pero tecnicamente no es exacto.

ademas solo tenemos 9 nodos en x, pero las funciones de graficacion dibujan
superficies muy suaves. eso puede dar la impresion de que tenemos un monton de
datos cuando en realidad la malla es bastante gruesa.
