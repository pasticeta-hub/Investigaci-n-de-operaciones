# Probabilidades del tablero de colores
contamos con 4 estados de colores en orden horario: **Naranja (x)** → **Azul (y)** → **Verde (z)** → **Rojo (t)** → **Naranja (x)** → **Azul (y)**...

tiramos 2 monedas y avanzamos en sentido horario:
- 0 caras → 0 pasos, prob. **1/4**
- 1 cara  → 1 paso,  prob. **1/2**
- 2 caras → 2 pasos, prob. **1/4**

Queremos calcular la evolución de las probabilidades \(X_n=(x_n,y_n,z_n,t_n)\) usando las fórmulas de transición vistas en clase.


```python
import matplotlib.pyplot as plt


x, y, z, t = 1.0, 0.0, 0.0, 0.0   # empezamos en naranja (creo jajsaj)
n = 25

x_i, y_i, z_i, t_i = [], [], [], []

for _ in range(n):
    # fórmulas de transición
    nx = (1/4)*x + 0*y     + (1/4)*z + (2/4)*t
    ny = (2/4)*x + (1/4)*y + 0*z     + (1/4)*t
    nz = (1/4)*x + (2/4)*y + (1/4)*z + 0*t
    nt = 0*x     + (1/4)*y + (2/4)*z + (1/4)*t

    # actualizamos variables
    x_i.append(nx); y_i.append(ny); z_i.append(nz); t_i.append(nt)
    x, y, z, t = nx, ny, nz, nt

print("Último vector X_n ≈", round(x,4), round(y,4), round(z,4), round(t,4),
      "   suma =", round(x+y+z+t,4))
```

Pimero que nada fijamos el estado inicial con `arranque` y el número de pasos `n` y luego colocamos \(X_0=(x,y,z,t)\) según el color elegido (o uniforme) y guardamos ese valor en

`x_i, y_i, z_i, t_i`.

Luego iteramos `n` veces aplicando las ecuaciones de transición (tal cual en clase):
  \[
  \begin{aligned}
  x' &= \tfrac14 x + 0\,y + \tfrac14 z + \tfrac24 t \\
  y' &= \tfrac24 x + \tfrac14 y + 0\,z + \tfrac14 t \\
  z' &= \tfrac14 x + \tfrac24 y + \tfrac14 z + 0\,t \\
  t' &= 0\,x + \tfrac14 y + \tfrac24 z + \tfrac14 t
  \end{aligned}
  \]
En cada vuelta del `for`: calculamos \(x',y',z',t'\), actualizamos \((x,y,z,t)\) y volvemos a guardar en las listas.
Al terminar, mostramos \(X_n\) con seis decimales y verificamos que la suma sea ≈ 1.


```python

plt.plot(x_i, label="Naranja")
plt.plot(y_i, label="Azul")
plt.plot(z_i, label="Verde")
plt.plot(t_i, label="Rojo")
plt.xlabel("n")
plt.ylabel("Probabilidad")
plt.title("Evolución de probabilidades (versión corta)")
plt.legend(); plt.grid(True, alpha=0.3)
plt.show()
```

ya por ultimo solo tomamos las series  `x_i, y_i, z_i, t_i` que guardamos en el bloque anterior, luego graficamos cada una, Etiquetamos ejes, ponemos título y leyenda, activamos una rejilla ligera para leer mejor los valores y mostramos la figura con `plt.show()`.
